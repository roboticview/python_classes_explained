```python
# create class room with lenght and with attributes
# and  method to calculate the area of room
```

```python
class Room:
    def __init__(self,length,width):
        self.length = length
        self.width = width
        self.area = self.area()

    def area(self):
        self.area = self.length * self.width
        return self.area
```

```python
room1 = Room(5,6)
```

```python
room2 = Room(7,8)
```

```python
floor = room1.area + room2.area
```
